```python
import speech_recognition as sr
import webbrowser
import os
import pyttsx3
import datetime

r = sr.Recognizer()
engine = pyttsx3.init()

with sr.Microphone() as source:
    print("Speak command...")
    r.adjust_for_ambient_noise(source)
    audio = r.listen(source)
    print("Listening complete...")

try:
    command = r.recognize_google(audio)
    print("You said:", command)

    command = command.lower()

    if "youtube" in command:
        print("Opening YouTube")
        engine.say("Opening YouTube")
        engine.runAndWait()
        webbrowser.open("https://youtube.com")

    elif "google" in command:
        print("Opening Google")
        engine.say("Opening Google")
        engine.runAndWait()
        webbrowser.open("https://google.com")

    elif "notepad" in command:
        print("Opening Notepad")
        engine.say("Opening Notepad")
        engine.runAndWait()
        os.system("notepad")

    elif "camera" in command:
        print("Opening Camera")
        engine.say("Opening Camera")
        engine.runAndWait()
        os.system("start microsoft.windows.camera:")

    elif "whatsapp" in command:
        print("Opening WhatsApp")
        engine.say("Opening WhatsApp")
        engine.runAndWait()
        webbrowser.open("https://web.whatsapp.com")

    elif "time" in command:
        current_time = datetime.datetime.now().strftime("%H:%M:%S")
        print("Time:", current_time)
        engine.say("The time is " + current_time)
        engine.runAndWait()

    elif "date" in command:
        today = str(datetime.date.today())
        print("Date:", today)
        engine.say("Today's date is " + today)
        engine.runAndWait()

    else:
        print("Command not recognized")
        engine.say("Sorry, I didn't understand that")
        engine.runAndWait()

except sr.UnknownValueError:
    print("Sorry, could not understand audio")
    engine.say("Sorry, I could not understand")
    engine.runAndWait()

except sr.RequestError:
    print("Internet connection issue")
    engine.say("Please check your internet connection")
    engine.runAndWait()
```
